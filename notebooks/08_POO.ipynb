<h2 id="poo">1. Programación Orientada a Objetos</h2>

Existen diferentes **paradigmas de programación**. Hasta este punto del curso, has trabajado de forma **procedimental**, es decir, siguiendo instrucciones paso a paso de arriba hacia abajo.

La **Programación Orientada a Objetos (POO)** es una solución potente para resolver problemas de organización de código y manejo de entidades complejas. Antes de sumergirnos en ella, observemos cómo se ve el enfoque tradicional.

```python
# Enfoque procedimental básico: paso a paso
name = input("Name: ")
house = input("House: ")
print(f"{name} from {house}")
```

### Abstracción mediante Funciones

Podemos mejorar el programa anterior utilizando funciones para "abstraer" la lógica. Esto hace que nuestra función `main` sea más legible y delegue tareas específicas a funciones dedicadas. 

Observa cómo `get_name` y `get_house` se encargan de los detalles de entrada, permitiendo que `main` se centre en la coordinación del flujo.

```python
def main():
    name = get_name()
    house = get_house()
    print(f"{name} from {house}")

def get_name():
    return input("Name: ")

def get_house():
    return input("House: ")

if __name__ == "__main__":
    main()
```

### Almacenamiento de datos en Tuplas

Podemos simplificar aún más nuestro programa almacenando al estudiante como una **Tupla**. Una tupla es una secuencia de valores. A diferencia de una lista, una tupla es **inmutable** (no se puede modificar). 

En esencia, estamos devolviendo dos valores bajo un mismo nombre. Observa cómo `get_student` empaqueta `name` y `house` en una sola entidad.

```python
def main():
    # Desempaquetado de la tupla (Unpacking)
    name, house = get_student()
    print(f"{name} from {house}")

def get_student():
    name = input("Name: ")
    house = input("House: ")
    return name, house # Devolvemos los dos valores juntos

if __name__ == "__main__":
    main()
```

### Indexación y Claridad Explicita

Podemos recibir la tupla completa en una sola variable llamada `student` y acceder a sus elementos mediante índices (`[0]`, `[1]`). 

Es una buena práctica usar paréntesis `(name, house)` al retornar para indicar explícitamente a cualquier programador que lea nuestro código que estamos devolviendo una tupla.

```python
def main():
    student = get_student()
    # Accedemos por índice
    print(f"{student[0]} from {student[1]}")

def get_student():
    name = input("Name: ")
    house = input("House: ")
    return (name, house) # Paréntesis para mayor claridad

if __name__ == "__main__":
    main()
```

### Inmutabilidad y Programación Defensiva

Las tuplas son **inmutables**, lo que significa que no podemos cambiar sus valores una vez creadas. La inmutabilidad es una herramienta de **programación defensiva**: nos asegura que los datos no cambiarán accidentalmente en otras partes del programa.



El siguiente código intentará cambiar la casa del estudiante si su nombre es "Padma", pero **producirá un error** porque Python no permite la reasignación en tuplas.

```python
def main():
    student = get_student()
    if student[0] == "Padma":
        # ¡ERROR! Las tuplas no soportan asignación de ítems
        student[1] = "Ravenclaw" 
    print(f"{student[0]} from {student[1]}")

def get_student():
    name = input("Name: ")
    house = input("House: ")
    return name, house

if __name__ == "__main__":
    main()
```

### Flexibilidad vs. Seguridad: El uso de Listas

Si quisiéramos permitir que otros programadores modifiquen los datos (por ejemplo, para corregir una casa mal asignada), podríamos usar una **Lista** en lugar de una tupla. 

A diferencia de las tuplas, las listas son **mutables**. Esto otorga flexibilidad, pero tiene un costo: la seguridad del código. Si el orden de los valores cambia o alguien modifica un dato por error, el programa podría comportarse de forma inesperada "río abajo".

```python
def main():
    student = get_student()
    # Como las listas son mutables, esto SÍ funciona:
    if student[0] == "Padma":
        student[1] = "Ravenclaw"
    print(f"{student[0]} from {student[1]}")

def get_student():
    name = input("Name: ")
    house = input("House: ")
    return [name, house] # Retornamos una lista (mutable)

if __name__ == "__main__":
    main()
```

### Diccionarios: Claridad mediante Pares Clave-Valor

Para evitar depender de índices numéricos (como `[0]` o `[1]`), que pueden ser confusos, podemos utilizar un **Diccionario**. Los diccionarios nos permiten asociar una **clave** (key) con un **valor** (value).

La gran ventaja aquí es la legibilidad: podemos acceder a los datos usando nombres descriptivos como `'name'` y `'house'`.

```python
def main():
    student = get_student()
    # Acceso mucho más claro mediante claves
    print(f"{student['name']} from {student['house']}")

def get_student():
    student = {}
    student["name"] = input("Name: ")
    student["house"] = input("House: ")
    return student

if __name__ == "__main__":
    main()
```

### Optimizando la creación del Diccionario

Podemos refinar aún más nuestra función `get_student`. En lugar de crear un diccionario vacío y luego asignar valores, podemos utilizar las llaves `{}` directamente en la sentencia `return` para crear y devolver el diccionario en una sola línea. 

Esto reduce el ruido visual y hace que el código sea más "pythónico".

```python
def main():
    student = get_student()
    print(f"{student['name']} from {student['house']}")

def get_student():
    name = input("Name: ")
    house = input("House: ")
    # Retorno directo usando un diccionario literal
    return {"name": name, "house": house}

if __name__ == "__main__":
    main()
```

### Mutabilidad y Legibilidad en Diccionarios

Al igual que las listas, los diccionarios son **mutables**. Podemos aprovechar esta característica para manejar casos especiales (como corregir la casa de un estudiante) utilizando los nombres de las claves en lugar de índices numéricos.

Esto hace que el código sea mucho más "auto-documentado", ya que cualquier programador entiende inmediatamente qué dato se está modificando al leer `student["house"]`.

```python
def main():
    student = get_student()
    # Manejamos el caso especial usando claves descriptivas
    if student["name"] == "Padma":
        student["house"] = "Ravenclaw"
    print(f"{student['name']} from {student['house']}")

def get_student():
    name = input("Name: ")
    house = input("House: ")
    # Retorno directo usando un diccionario literal
    return {"name": name, "house": house}

if __name__ == "__main__":
    main()
```

<h2 id="clases">2. Clases</h2>

En la programación orientada a objetos, las **clases** son una forma de crear nuestros propios tipos de datos y darles un nombre descriptivo.

Imagina que una clase es como un **molde** (blueprint) para un tipo de datos: inventamos el nombre del tipo y definimos qué información debe contener.


### Implementación Inicial de la Clase `Student`

Comencemos definiendo una clase muy simple. Por convención en Python, los nombres de las clases comienzan con **mayúscula** (CapWords). El uso de `...` (Ellipsis) es una forma de decirle a Python: "aquí va algo, pero lo definiré más tarde".



```python
class Student:
    ...

def main():
    student = get_student()
    # Usamos la "notación de punto" (.) para acceder a los atributos
    print(f"{student.name} from {student.house}")

def get_student():
    # Instanciamos un objeto de la clase Student
    student = Student()
    student.name = input("Name: ")
    student.house = input("House: ")
    return student

if __name__ == "__main__":
    main()
```

### Objetos e Instancias

Cada vez que utilizas el "plano" de una clase para crear algo concreto, estás creando lo que llamamos un **"objeto"** o una **"instancia"**. 

* **Clase:** `Student` (El concepto/molde).
* **Objeto:** `student` (La realización concreta con datos específicos).

En el código anterior, utilizamos la **"notación de punto"** (`student.name`) para asignar y acceder a los atributos de ese objeto específico.

### El Método `__init__` y el concepto de `self`

Podemos estandarizar qué atributos debe tener cada objeto de la clase `Student` desde el momento de su creación. Para ello usamos una función especial dentro de la clase llamada **método**.

El método `__init__` es el **constructor**. Se ejecuta automáticamente cuando creas un nuevo objeto:

* **`self`**: Es una referencia al objeto que se acaba de crear. Permite que el objeto guarde sus propios datos.
* **Atributos**: Al hacer `self.name = name`, estamos guardando el valor pasado como argumento dentro de la "mochila" del objeto.

```python
class Student:
    # El constructor estandariza los atributos del objeto
    def __init__(self, name, house):
        self.name = name
        self.house = house

def main():
    student = get_student()
    print(f"{student.name} from {student.house}")

def get_student():
    name = input("Name: ")
    house = input("House: ")
    # Llamamos al constructor pasando los argumentos necesarios
    student = Student(name, house)
    return student

if __name__ == "__main__":
    main()
```

### Simplificando la creación del objeto

Al igual que hicimos con los diccionarios, podemos simplificar el código eliminando variables intermedias innecesarias. Podemos llamar al constructor `Student(name, house)` directamente en la sentencia `return`.

```python
class Student:
    def __init__(self, name, house):
        self.name = name
        self.house = house

def main():
    student = get_student()
    print(f"{student.name} from {student.house}")

def get_student():
    name = input("Name: ")
    house = input("House: ")
    # Retorno directo de la nueva instancia
    return Student(name, house)

if __name__ == "__main__":
    main()
```

<h2 id="raise">3. raise</h2>

La programación orientada a objetos nos anima a encapsular toda la funcionalidad de una entidad dentro de su propia definición. Pero, **¿qué pasa si algo sale mal?** ¿Qué pasa si alguien intenta crear un estudiante sin nombre o asignarle una casa que no existe (como "Calle Falsa 123")?

Podemos utilizar la palabra clave **`raise`** para lanzar excepciones propias y alertar al programador sobre errores cometidos por el usuario. En este caso, usaremos `ValueError` con mensajes específicos.


```python
class Student:
    def __init__(self, name, house):
        # Validación de integridad
        if not name:
            raise ValueError("Falta el nombre")
        if house not in ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]:
            raise ValueError("Casa inválida")
        
        self.name = name
        self.house = house

def main():
    try:
        student = get_student()
        print(f"{student.name} from {student.house}")
    except ValueError as e:
        print(f"Error: {e}")

def get_student():
    name = input("Name: ")
    house = input("House: ")
    return Student(name, house)

if __name__ == "__main__":
    main()
```

### Representación de Objetos: El método `__str__`

Python permite definir una función específica para determinar qué se imprime cuando pasamos un objeto a la función `print()`. Este es el método mágico **`__str__`**.

Sin este método, si intentas imprimir un objeto, verás algo críptico como `<__main__.Student object at 0x...>`. Al implementarlo, puedes decidir exactamente qué atributos mostrar.

```python
class Student:
    def __init__(self, name, house, patronus):
        if not name:
            raise ValueError("Falta el nombre")
        if house not in ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]:
            raise ValueError("Casa inválida")
        self.name = name
        self.house = house
        self.patronus = patronus

    # Método mágico para representación en string
    def __str__(self):
        return f"{self.name} de {self.house}"

def main():
    student = get_student()
    # Ahora podemos imprimir el objeto directamente
    print(student)

def get_student():
    name = input("Name: ")
    house = input("House: ")
    patronus = input("Patronus: ")
    return Student(name, house, patronus)

if __name__ == "__main__":
    main()
```

### Métodos Personalizados: Comportamiento de los Objetos

`__str__` es un método incorporado, ¡pero podemos crear nuestros propios métodos! A diferencia de los diccionarios, las clases pueden contener funciones que definen el comportamiento de los objetos.

Vamos a crear un método llamado `charm` (encantamiento) que devuelva un emoji basado en el patronus del estudiante.

```python
class Student:
    def __init__(self, name, house, patronus=None):
        if not name:
            raise ValueError("Falta el nombre")
        if house not in ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]:
            raise ValueError("Casa inválida")
        self.name = name
        self.house = house
        self.patronus = patronus

    def __str__(self):
        return f"{self.name} de {self.house}"

    # Nuestro propio método
    def charm(self):
        match self.patronus:
            case "Ciervo":
                return "🦌"
            case "Nutria":
                return "🦦"
            case "Jack Russell terrier":
                return "🐶"
            case _:
                return "🪄"

def main():
    student = get_student()
    print("¡Expecto Patronum!")
    print(student.charm())

def get_student():
    name = input("Name: ")
    house = input("House: ")
    patronus = input("Patronus: ") or None
    return Student(name, house, patronus)

if __name__ == "__main__":
    main()
```

### Preparando el terreno: Limpieza de Código

Antes de avanzar hacia los Decoradores, simplificaremos nuestra clase `Student` eliminando el código del patronus para centrarnos en los conceptos básicos de seguridad de atributos.

Observa que, aunque tenemos validación en el `__init__`, todavía hay un problema: **nada impide que alguien cambie la casa manualmente** después de crear el objeto, saltándose las reglas.

```python
class Student:
    def __init__(self, name, house):
        if not name:
            raise ValueError("Nombre inválido")
        if house not in ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]:
            raise ValueError("Casa inválida")
        self.name = name
        self.house = house

    def __str__(self):
        return f"{self.name} de {self.house}"

def main():
    student = get_student()
    # Peligro: Estamos cambiando la casa a una inválida manualmente
    student.house = "Calle Falsa 123" 
    print(student)

def get_student():
    name = input("Name: ")
    house = input("House: ")
    return Student(name, house)

if __name__ == "__main__":
    main()
```

<h2 id="decoradores">4. Decoradores (Decorators) y Propiedades</h2>

Los **Decoradores** en Python son funciones que modifican el comportamiento de otras funciones y se identifican porque comienzan con el símbolo `@`. 

Para blindar nuestro código, utilizamos decoradores para definir **propiedades**. Esto nos permite interceptar el momento en que alguien intenta leer un atributo (Getter) o modificarlo (Setter), permitiéndonos ejecutar validaciones en tiempo real.

```python
class Student:
    def __init__(self, name, house):
        if not name:
            raise ValueError("Nombre inválido")
        # Aquí el constructor llamará automáticamente al setter de 'house'
        self.name = name
        self.house = house

    def __str__(self):
        return f"{self.name} de {self.house}"

    # Getter para house: se llama cuando alguien hace 'print(student.house)'
    @property
    def house(self):
        return self._house

    # Setter para house: se llama cuando alguien hace 'student.house = "Gryffindor"'
    @house.setter
    def house(self, house):
        if house not in ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]:
            raise ValueError("Casa inválida")
        self._house = house

def main():
    student = get_student()
    print(student)

def get_student():
    name = input("Name: ")
    house = input("House: ")
    return Student(name, house)

if __name__ == "__main__":
    main()
```

### ¿Por qué `_house` y no `house`?

Habrás notado que dentro de la clase usamos `self._house` (con un guion bajo). Aquí la explicación:

1. **`house`**: Es la **propiedad**. Es el nombre "público" con el que el usuario interactúa.
2. **`_house`**: Es el **atributo real** donde se guarda el dato. 

**La convención del guion bajo (`_`)**: En Python, un guion bajo al principio de una variable indica a otros programadores: *"Oye, este es un detalle interno, por favor no lo toques directamente"*. 

Si intentáramos usar `self.house = house` dentro del setter, entraríamos en un bucle infinito (recursión), porque el setter se llamaría a sí mismo eternamente. Por eso usamos `_house` para guardar el valor final una vez validado.

### Protegiendo todos los flancos: Validación del Nombre

Al igual que protegimos la casa, podemos aplicar la misma lógica al nombre del estudiante. De esta forma, nos aseguramos de que el nombre nunca sea una cadena vacía, no solo en el momento de crear el objeto (`__init__`), sino en cualquier momento de la vida del objeto.

```python
class Student:
    def __init__(self, name, house):
        # Al asignar aquí, se disparan los @setters correspondientes
        self.name = name
        self.house = house

    def __str__(self):
        return f"{self.name} de {self.house}"

    # Getter y Setter para Name
    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, name):
        if not name:
            raise ValueError("Nombre inválido")
        self._name = name

    # Getter y Setter para House
    @property
    def house(self):
        return self._house

    @house.setter
    def house(self, house):
        if house not in ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]:
            raise ValueError("Casa inválida")
        self._house = house

def main():
    student = get_student()
    print(f"Estudiante cargado: {student}")

def get_student():
    name = input("Name: ")
    house = input("House: ")
    return Student(name, house)

if __name__ == "__main__":
    main()
```

<h2 id="metodos-clase">6. Métodos de Clase</h2>

A veces, queremos añadir funcionalidad a una clase que no dependa de una instancia u objeto específico, sino de la **clase como un todo**. Para esto utilizamos el decorador **`@classmethod`**.

Mientras que los métodos de instancia reciben `self` (el objeto específico), los métodos de clase reciben **`cls`** (la clase misma).



### Ejemplo 1: El Sombrero Seleccionador (Sin Método de Clase)

Imagina que queremos simular el Sombrero Seleccionador. En una primera aproximación, podríamos crear una instancia del sombrero para usar su funcionalidad.

```python
import random

class Hat:
    def __init__(self):
        # La lista de casas pertenece a la instancia
        self.houses = ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]

    def sort(self, name):
        print(name, "está en", random.choice(self.houses))

# Tenemos que crear un objeto 'hat' para poder usar 'sort'
hat = Hat()
hat.sort("Harry")
```

### Ejemplo 2: Refactorización con `@classmethod`

Realmente solo existe **un** Sombrero Seleccionador. No tiene mucho sentido crear instancias de él si la funcionalidad siempre es la misma. 

Podemos mover la lista `houses` para que sea una **variable de clase** (compartida por todos) y convertir `sort` en un **método de clase**. Observa que:
1. Eliminamos el `__init__` (ya no necesitamos instanciarlo).
2. Cambiamos `self` por `cls`.
3. Llamamos al método directamente usando el nombre de la clase: `Hat.sort()`.

```python
import random

class Hat:
    # Variable de clase: compartida por toda la clase Hat
    houses = ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]

    @classmethod
    def sort(cls, name):
        # Accedemos a la variable de clase usando 'cls'
        print(name, "está en", random.choice(cls.houses))

# Llamada directa a la clase, sin crear un objeto previo
Hat.sort("Harry")
```

### Aplicación Práctica: El Patrón Factory en `Student`

Podemos aplicar este concepto a nuestra clase `Student` para "limpiar" nuestro código principal. En lugar de tener una función global `get_student`, podemos crear un método de clase dentro de `Student` llamado `get`.

Este método se encarga de solicitar los datos y **devolver una instancia de la propia clase** llamando a `cls(name, house)`. Este es un patrón de diseño común llamado *Factory* (Fábrica).

```python
class Student:
    def __init__(self, name, house):
        self.name = name
        self.house = house

    def __str__(self):
        return f"{self.name} de {self.house}"

    @classmethod
    def get(cls):
        # El método de clase se encarga de su propia creación
        name = input("Name: ")
        house = input("House: ")
        return cls(name, house)

def main():
    # Llamamos al método de clase para obtener un objeto Student
    student = Student.get()
    print(student)

if __name__ == "__main__":
    main()
```

<h2 id="metodos-estaticos">7. Métodos Estáticos</h2>

Además de los métodos de instancia (que usan `self`) y los métodos de clase (que usan `cls`), Python ofrece los **Métodos Estáticos** mediante el decorador **`@staticmethod`**.

### ¿Qué es un Método Estático?
Es una función que vive dentro de la clase por razones de organización, pero que **no necesita acceder** ni a los datos del objeto (`self`) ni a los datos de la clase (`cls`). Se comporta como una función normal, pero bajo el espacio de nombres de la clase.



| Tipo de Método | Decorador | Primer Argumento | Acceso a... |
| :--- | :--- | :--- | :--- |
| **Instancia** | Ninguno | `self` | Atributos del objeto específico |
| **Clase** | `@classmethod` | `cls` | Variables de la clase / Otros métodos de clase |
| **Estático** | `@staticmethod` | Ninguno | Nada (es una función de utilidad) |

### Ejemplo Práctico: Utilidades de Validación

Imagina que en nuestra clase `Student` queremos una función que verifique si una cadena de texto es un código de estudiante válido (por ejemplo, que empiece con "H" de Hogwarts seguido de números). 

Esta función no necesita saber quién es el estudiante, ni cuántos estudiantes hay; es simplemente una herramienta de utilidad relacionada con la lógica de los estudiantes.

```python
class Student:
    def __init__(self, name, house):
        self.name = name
        self.house = house

    @staticmethod
    def is_valid_id(student_id):
        # Función de utilidad: no usa self ni cls
        # Solo verifica un formato: "H" seguido de 3 dígitos
        if len(student_id) == 4 and student_id.startswith("H"):
            return student_id[1:].isdigit()
        return False

# Podemos llamar al método estático sin crear un estudiante
print(f"¿Es H123 válido?: {Student.is_valid_id('H123')}")
print(f"¿Es A500 válido?: {Student.is_valid_id('A500')}")

# También se puede llamar desde una instancia, pero no es lo común
harry = Student("Harry", "Gryffindor")
print(f"¿Harry tiene un ID válido?: {harry.is_valid_id('H999')}")
```

### ¿Cuándo usar `@staticmethod`?

Se recomienda usarlo cuando tienes una función que está estrechamente relacionada con la clase, pero que podría existir fuera de ella sin problemas. 

**Ventaja:** Mejora la organización del código y comunica a otros programadores que ese método no modificará el estado del objeto ni de la clase, lo que lo hace más seguro y fácil de testear.

<h2 id="herencia">8. Herencia</h2>

La **Herencia** es, quizás, la característica más poderosa de la programación orientada a objetos. Permite crear una clase que "hereda" métodos, variables y atributos de otra clase.

### Conceptos Clave:
* **Clase Padre (Superclase):** La clase de la que se hereda (la más general).
* **Clase Hija (Subclase):** La clase que hereda (la más específica).
* **Relación "Es-un":** Un Estudiante **es un** Mago. Un Profesor **es un** Mago.



### Implementación Base: `Wizard`, `Student` y `Professor`

En el mundo de Harry Potter, tanto los estudiantes como los profesores tienen nombres y son magos. En lugar de repetir la validación del nombre en ambas clases, la centralizamos en una superclase llamada `Wizard`.

```python
class Wizard:
    def __init__(self, name):
        # La validación del nombre ocurre una sola vez aquí
        if not name:
            raise ValueError("Falta el nombre")
        self.name = name

class Student(Wizard):
    def __init__(self, name, house):
        # super() accede a la clase padre (Wizard) y llama a su __init__
        super().__init__(name)
        self.house = house

class Professor(Wizard):
    def __init__(self, name, subject):
        # También heredamos la lógica de Wizard
        super().__init__(name)
        self.subject = subject

# Instanciación
wizard = Wizard("Albus")
student = Student("Harry", "Gryffindor")
professor = Professor("Severus", "Defensa Contra las Artes Oscuras")

print(f"{wizard.name} es un mago genérico.")
print(f"{student.name} es un estudiante de {student.house}.")
print(f"{professor.name} enseña {professor.subject}.")
```

### El Poder de `super()` y el Principio DRY

La función **`super()`** es fundamental. Permite que la clase hija llame al constructor de la clase padre, evitando que tengamos que volver a escribir `self.name = name` y sus validaciones en cada nueva clase que creemos.

Esto cumple con el principio **DRY (Don't Repeat Yourself)**: si mañana decidimos que los nombres no pueden tener números, solo tenemos que cambiar la validación en `Wizard` y automáticamente se aplicará a `Student`, `Professor` y cualquier otra clase futura.

### Ejemplo Expandido: Métodos Heredados

No solo heredamos atributos, también heredamos **métodos**. Observa cómo definimos un comportamiento general en la clase padre que todas las hijas pueden usar.

```python
class Mago:
    def __init__(self, nombre):
        self.nombre = nombre
    
    def presentarse(self):
        return f"Hola, mi nombre es {self.nombre}."

class Estudiante(Mago):
    def __init__(self, nombre, casa):
        super().__init__(nombre)
        self.casa = casa
    
    # Podemos extender el comportamiento
    def presentarse(self):
        # Usamos el saludo general y le añadimos algo específico
        saludo_base = super().presentarse()
        return f"{saludo_base} Y soy de la casa {self.casa}."

class Profesor(Mago):
    def __init__(self, nombre, materia):
        super().__init__(nombre)
        self.materia = materia

# Pruebas
ron = Estudiante("Ron", "Gryffindor")
snape = Profesor("Severus", "Pociones")

# El estudiante usa su versión extendida de presentarse
print(ron.presentarse())

# El profesor usa la versión original de la clase padre (Mago)
# porque no la sobreescribimos.
print(snape.presentarse())
```

### 💡 Resumen de Herencia
1. **Reutilización:** No repetimos código de validación o asignación.
2. **Jerarquía:** Creamos estructuras lógicas (Mago -> Estudiante).
3. **Mantenimiento:** Los cambios en la clase padre se propagan a las hijas.

<h2 id="herencia-excepciones">9. Herencia y Excepciones</h2>

Aunque acabamos de introducir la herencia de forma teórica, la has estado utilizando desde que aprendiste a manejar errores. En Python, las excepciones no son simples cadenas de texto; son **clases** organizadas en una jerarquía de padres, hijos y abuelos.

### La Jerarquía de Excepciones

Todas las excepciones derivan de una clase base. Si capturas una "clase padre", también capturarás a todas sus "clases hijas". 



Aquí un vistazo a la estructura simplificada:
* **BaseException** (El tatarabuelo)
    * **Exception** (El abuelo de la mayoría de errores comunes)
        * **ArithmeticError** (Padre de errores matemáticos)
            * `ZeroDivisionError` (Hijo)
        * **LookupError** (Padre de errores de búsqueda)
            * `KeyError` (Hijo: falta una clave en un dict)
            * `IndexError` (Hijo: índice fuera de rango en lista)
        * **ValueError** (Hijo de Exception)


### ¿Por qué es importante esto?

Gracias a la herencia, podemos ser tan específicos o tan generales como queramos al capturar errores. 

Si intentamos capturar `Exception`, atraparemos casi cualquier error que ocurra (porque casi todos heredan de ella). Pero la buena práctica dicta ser específicos para no ocultar errores inesperados.

```python
# Ejemplo de cómo la herencia afecta al manejo de errores

def dividir_numeros(a, b):
    return a / b

try:
    # Caso 1: División por cero
    resultado = dividir_numeros(10, 0)
except ArithmeticError: 
    # ArithmeticError es el PADRE de ZeroDivisionError
    # Por lo tanto, este bloque capturará el error aunque no hayamos 
    # puesto explícitamente "ZeroDivisionError"
    print("Error: Se detectó un problema aritmético (posible división por cero).")

try:
    # Caso 2: Error de valor
    int("Hogwarts")
except Exception as e:
    # Exception es el PADRE de ValueError. 
    # Capturará CUALQUIER error que herede de Exception.
    print(f"Error capturado por la clase padre: {e}")
```

<h2 id="sobrecarga-operadores">10. Sobrecarga de Operadores</h2>

¿Alguna vez te has preguntado por qué el signo `+` sirve tanto para sumar números como para unir cadenas de texto? Eso es gracias a la **Sobrecarga de Operadores**. 

En Python, podemos definir métodos especiales (métodos mágicos) para que nuestros objetos respondan a operadores aritméticos tradicionales de una manera lógica para nuestra aplicación.



### Ejemplo: La Bóveda de Gringotts (`Vault`)

Imagina que queremos sumar el dinero de dos personas en el mundo de Harry Potter. El dinero se divide en Galeones, Sickles y Knuts. No son simples números, son objetos complejos.

```python
class Vault:
    def __init__(self, galleons=0, sickles=0, knuts=0):
        self.galleons = galleons
        self.sickles = sickles
        self.knuts = knuts

    def __str__(self):
        return f"{self.galleons} Galeons, {self.sickles} Sickles, {self.knuts} Knuts"

    # Sobrecarga del operador +
    def __add__(self, other):
        galleons = self.galleons + other.galleons
        sickles = self.sickles + other.sickles
        knuts = self.knuts + other.knuts
        # Devolvemos un NUEVO objeto Vault con los totales
        return Vault(galleons, sickles, knuts)

# Creamos la bóveda de Harry
potter = Vault(100, 50, 25)
print(f"Bóveda de Harry: {potter}")

# Creamos la bóveda de los Weasley
weasley = Vault(25, 50, 100)
print(f"Bóveda de los Weasley: {weasley}")

# ¡Sumamos los objetos directamente!
total = potter + weasley
print(f"Total combinado: {total}")
```

### ¿Cómo funciona `__add__`?

Cuando Python ve la expresión `potter + weasley`, lo traduce internamente a:
`potter.__add__(weasley)`

* **`self`**: Se refiere al objeto que está a la **izquierda** del operador (en este caso, `potter`).
* **`other`**: Se refiere al objeto que está a la **derecha** del operador (en este caso, `weasley`).

Este patrón se aplica a otros operadores como `__sub__` para la resta (`-`) o `__mul__` para la multiplicación (`*`). Esto hace que nuestro código sea increíblemente intuitivo de leer.
