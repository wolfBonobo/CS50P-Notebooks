<h2 id="random">2. M칩dulo Random</h2>

---
La librer칤a `random` viene integrada con Python y permite generar informaci칩n aleatoria. Para utilizar cualquier m칩dulo externo, debemos cargarlo usando la palabra clave **`import`**.

### A. Uso de `random.choice(seq)`
Esta funci칩n toma una **secuencia** (como una lista) y elige un elemento al azar.

````python
import random

# Python calcula las probabilidades autom치ticamente (50/50 en este caso)
coin = random.choice(["heads", "tails"])
print(coin)
````

### B. Mejorando el c칩digo con `from`
La palabra clave `from` nos permite ser espec칤ficos sobre qu칠 queremos importar de un m칩dulo. 

* **Eficiencia:** Solo cargamos la funci칩n necesaria en lugar de todo el m칩dulo, lo que puede ahorrar recursos del sistema.
* **Sintaxis:** Ya no es necesario usar el prefijo `random.`, podemos llamar a la funci칩n directamente por su nombre.

````python
from random import choice

# Ahora usamos 'choice' directamente sin el prefijo del m칩dulo
coin = choice(["heads", "tails"])
print(coin)
````

### C. `random.randint(a, b)`
Esta funci칩n genera un n칰mero entero aleatorio dentro de un rango inclusivo entre **a** y **b**.

### D. `random.shuffle(x)`
Esta funci칩n mezcla los elementos de una lista en un orden aleatorio.

> **Nota t칠cnica:** A diferencia de otras funciones, `shuffle` realiza los cambios **in-place** (en el lugar). Esto significa que no devuelve una nueva lista, sino que modifica la lista original directamente.

````python
import random

# Generar un n칰mero entre 1 y 10
number = random.randint(1, 10)
print(f"N칰mero aleatorio: {number}")

# Mezclar una lista de cartas
cards = ["jack", "queen", "king"]
random.shuffle(cards)

for card in cards:
    print(card)
````

<h2 id="cli">4. Argumentos de L칤nea de Comandos (CLI Arguments)</h2>

---
Hasta ahora, hemos proporcionado valores dentro del c칩digo o mediante `input()`. Los **argumentos de l칤nea de comandos** nos permiten pasar informaci칩n al programa en el momento exacto de su ejecuci칩n.

### El m칩dulo `sys`
El m칩dulo `sys` (system) provee funciones y variables que interact칰an con el int칠rprete de Python.

* **`sys.argv`**: Es una **lista** que registra todo lo que el usuario escribe en la terminal al ejecutar el programa.

### Anatom칤a de `sys.argv`
Imagina que ejecutas: `python name.py bonobo`

| 칈ndice | Contenido | Descripci칩n |
| :--- | :--- | :--- |
| `sys.argv[0]` | `name.py` | El nombre del archivo que se est치 ejecutando. |
| `sys.argv[1]` | `bonobo` | El primer argumento proporcionado por el usuario. |

````python
import sys

# Intentamos imprimir el argumento en la posici칩n 1
# Nota: Si ejecutas esto sin argumentos, fallar치 con un IndexError
try:
    print("hello, my name is", sys.argv[1])
except IndexError:
    print("Too few arguments")
````

### Mejorando la robustez
Podemos ser m치s espec칤ficos y defensivos validando la longitud de la lista `sys.argv` antes de intentar acceder a sus elementos. Esto evita que el programa "explote" y ofrece feedback claro al usuario.

### La funci칩n `sys.exit`
`sys.exit()` nos permite detener el programa inmediatamente. Es ideal para el manejo de errores porque separa la l칩gica de validaci칩n del resto del c칩digo. Si ocurre un error, salimos; si no, el c칩digo contin칰a limpiamente.

````python
import sys

# Comprobamos la cantidad de argumentos
if len(sys.argv) < 2:
    sys.exit("Too few arguments") # Detiene el programa con este mensaje
elif len(sys.argv) > 2:
    sys.exit("Too many arguments")

# Si el c칩digo llega aqu칤, sabemos que sys.argv[1] existe de forma segura
print("hello, my name is", sys.argv[1])
````

<h2 id="slice">5. Slicing (Rebanado de Listas)</h2>

---
El **slicing** es una t칠cnica que nos permite tomar una lista y decirle al int칠rprete exactamente d칩nde queremos que comience y d칩nde queremos que termine la selecci칩n.

### El Problema de la Iteraci칩n Completa
Si intentamos iterar sobre la lista completa de argumentos (`sys.argv`), el programa incluir치 el nombre del archivo como si fuera un dato m치s del usuario.

````python
import sys

# Si ejecutamos: python name.py bonobo
# El resultado incluir치: hello, my name is name.py
if len(sys.argv) < 2:
    sys.exit("Too few arguments")

for arg in sys.argv:
    print("hello, my name is", arg)
````

### La Soluci칩n: Definir rangos con `[:]`
Podemos usar corchetes y el operador `:` para indicar un subconjunto de la lista. 

* **`sys.argv[1:]`**: Le indica al int칠rprete que empiece en el 칤ndice **1** y contin칰e hasta el final de la lista.
* De esta manera, ignoramos el 칤ndice **0** donde se almacena el nombre del programa (`name.py`).

````python
import sys

if len(sys.argv) < 2:
    sys.exit("Too few arguments")

# Usamos slicing para ignorar el primer elemento (index 0)
for arg in sys.argv[1:]:
    print("hello, my name is", arg)
````

<h2 id="packages">6. Paquetes (Packages)</h2>

---
Una de las razones por las que Python es tan popular es la enorme cantidad de librer칤as de terceros que a침aden funcionalidades incre칤bles. Implementadas como carpetas, estas librer칤as externas se denominan **paquetes**.

* **PyPI (Python Package Index):** Es el repositorio o directorio central donde se encuentran todos los paquetes de terceros disponibles actualmente.
* **pip:** Es el gestor de paquetes de Python que te permite instalar estas librer칤as r치pidamente en tu sistema.



### Instalaci칩n de Paquetes
Para instalar un paquete, se utiliza la terminal. Por ejemplo, para instalar el famoso paquete **`cowsay`** (que permite que una vaca "hable" al usuario), escribir칤amos:
```bash

```
### Uso de paquetes en el c칩digo: `say.py`
Una vez que el paquete est치 instalado en el sistema, podemos importarlo en nuestro archivo `.py` como cualquier otro m칩dulo. 

En este ejemplo, utilizaremos la librer칤a `cowsay` combinada con `sys.argv` para capturar el nombre del usuario desde la terminal.

````python
import cowsay
import sys

# Comprobamos que el usuario introdujo exactamente un argumento (su nombre)
if len(sys.argv) == 2:
    # La funci칩n cow genera un dibujo de una vaca diciendo el texto
    cowsay.cow("hello, " + sys.argv[1])
````

### Explorando otras funciones del paquete
Los paquetes de terceros suelen incluir m칰ltiples funcionalidades o variaciones. En el caso de `cowsay`, no estamos limitados solo a vacas; podemos invocar otros personajes como un **T-Rex** simplemente cambiando la funci칩n que llamamos.

````python
import cowsay
import sys

if len(sys.argv) == 2:
    # Sustituimos la funci칩n cow por la funci칩n trex
    cowsay.trex("hello, " + sys.argv[1])

# Ejecuci칩n sugerida en terminal: python say.py bonobo
````

# El Formato JSON

Los servidores suelen responder con datos en formato **JSON** (*JavaScript Object Notation*), un est치ndar basado en texto dise침ado para intercambiar datos de forma sencilla entre aplicaciones. 



En Python, este formato se interpreta autom치ticamente convirti칠ndolo en un **diccionario**, lo que permite acceder a la informaci칩n de manera estructurada y eficiente.

---

### 游냀 Ejemplo en Python

Para trabajar con JSON en Python, se utiliza la librer칤a integrada `json`:

```python
import json

# Una cadena en formato JSON (texto)
json_data = '{"fruta": "manzana", "calorias": 130}'

# Convertir JSON a un diccionario de Python
datos = json.loads(json_data)

print(datos["fruta"])   # Salida: manzana
print(datos["calorias"]) # Salida: 130

### Ejemplo Pr치ctico: iTunes Search API
Apple ofrece una API p칰blica para buscar m칰sica. Podemos construir una URL espec칤fica (ej: `entity=song&limit=1&term=weezer`) para obtener informaci칩n sobre canciones en tiempo real. 

Al principio, el resultado puede parecer confuso y dif칤cil de leer porque es una estructura de datos muy densa.

````python
import requests
import sys

# Verificamos que el usuario introdujo un t칠rmino de b칰squeda
if len(sys.argv) != 2:
    sys.exit()

# Realizamos la petici칩n GET a la API de iTunes
# El t칠rmino de b칰squeda viene de los argumentos de la l칤nea de comandos
response = requests.get("https://itunes.apple.com/search?entity=song&limit=1&term=" + sys.argv[1])

# Mostramos el JSON crudo (convertido a diccionario por Python)
print(response.json())
````

### Mejorando la visualizaci칩n con `json.dumps`
Para entender la estructura de los datos que recibimos, Python incluye la librer칤a nativa `json`. La funci칩n `json.dumps` (con el argumento `indent`) nos permite "imprimir bonito" el resultado, facilitando la identificaci칩n de las llaves y valores.

````python
import json
import requests
import sys

if len(sys.argv) != 2:
    sys.exit()

response = requests.get("https://itunes.apple.com/search?entity=song&limit=1&term=" + sys.argv[1])

# json.dumps convierte el diccionario en un string formateado con sangr칤a (indent)
print(json.dumps(response.json(), indent=2))
````

### Navegando por los resultados
Una vez que vemos la estructura formateada, notamos que Apple devuelve un diccionario con una llave llamada `"results"`, que contiene una lista de canciones. Cada canci칩n es, a su vez, un diccionario con llaves como `"trackName"`.

Podemos iterar sobre esta lista para extraer solo los nombres de las canciones.

````python
import requests
import sys

if len(sys.argv) != 2:
    sys.exit()

# Aumentamos el l칤mite a 50 resultados
response = requests.get("https://itunes.apple.com/search?entity=song&limit=50&term=" + sys.argv[1])

# Guardamos el JSON en una variable (o de objetos)
o = response.json()

# Iteramos sobre la lista de resultados para imprimir solo el nombre de la pista
for result in o["results"]:
    print(result["trackName"])
````
