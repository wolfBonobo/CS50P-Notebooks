<h2 id="regex">1. Introducción a las Expresiones Regulares</h2>


Las expresiones regulares o "regexes" nos permitirán examinar patrones dentro de nuestro código. Por ejemplo, podríamos querer validar que una dirección de correo electrónico esté formateada correctamente. Las expresiones regulares nos permitirán examinar expresiones de esta manera.

Para comenzar, escribe `code validate.py` en la ventana de la terminal. Luego, codifica lo siguiente en el editor de texto:

```python
email = input("What's your email? ").strip()

if "@" in email:
    print("Valid")
else:
    print("Invalid")
```

Nota que `strip` eliminará los espacios en blanco al principio o al final de la entrada. Al ejecutar este programa, verás que mientras se ingrese un símbolo `@`, el programa considerará la entrada como válida.

Puedes imaginar, sin embargo, que alguien podría ingresar solo `@@` y la entrada podría ser considerada como válida. Podríamos considerar que una dirección de correo electrónico tiene al menos un `@` y un `.` en alguna parte de ella. Modifica tu código de la siguiente manera:

```python
email = input("What's your email? ").strip()

if "@" in email and "." in email:
    print("Valid")
else:
    print("Invalid")
```

Nota que, aunque esto funciona como se espera, nuestro usuario podría ser adversario, escribiendo simplemente `@.` resultaría en que el programa devuelva `valid`.

Podemos mejorar la lógica de nuestro programa de la siguiente manera:

```python
email = input("What's your email? ").strip()

username, domain = email.split("@")

if username and "." in domain:
    print("Valid")
else:
    print("Invalid")
```

Nota cómo el método `strip` (o en este caso `split`) se utiliza para determinar si `username` existe y si `.` está dentro de la variable `domain`. Al ejecutar este programa, una dirección de correo electrónico estándar ingresada por ti podría ser considerada `valid`. Al ingresar solo `malan@harvard`, encontrarás que el programa considera esta entrada como `invalid`.

Podemos ser aún más precisos, modificando nuestro código de la siguiente manera:

```python
email = input("What's your email? ").strip()

username, domain = email.split("@")

if username and domain.endswith(".edu"):
    print("Valid")
else:
    print("Invalid")
```

Nota cómo el método `endswith` comprobará si `domain` contiene `.edu`. Sin embargo, aún así, un usuario nefasto todavía podría romper nuestro código. Por ejemplo, un usuario podría escribir `malan@.edu` y sería considerado válido.

De hecho, podríamos seguir iterando sobre este código nosotros mismos. Sin embargo, resulta que Python tiene una biblioteca existente llamada `re` que tiene una serie de funciones integradas que pueden validar entradas de usuarios contra patrones.

Una de las funciones más versátiles dentro de la biblioteca `re` es `search`.

La función `search` sigue la firma `re.search(pattern, string, flags=0)`. Siguiendo esta firma, podemos modificar nuestro código de la siguiente manera:

```python
import re

email = input("What's your email? ").strip()

if re.search("@", email):
    print("Valid")
else:
    print("Invalid")
```

Nota que esto no aumenta en absoluto la funcionalidad de nuestro programa. De hecho, es de alguna manera un paso atrás.

Podemos avanzar más en la funcionalidad de nuestro programa. Sin embargo, necesitamos avanzar en nuestro vocabulario en torno a la validación. Resulta que en el mundo de las expresiones regulares hay ciertos símbolos que nos permiten identificar patrones. En este punto, solo hemos estado verificando piezas específicas de texto como `@`. Sucede que muchos símbolos especiales pueden pasarse al intérprete con el propósito de participar en la validación. Una lista no exhaustiva de esos patrones es la siguiente:

* `.`   cualquier carácter excepto una nueva línea
* `*`   0 o más repeticiones
* `+`   1 o más repeticiones
* `?`   0 o 1 repetición
* `{m}` m repeticiones
* `{m,n}` m-n repeticiones

Implementando esto dentro de nuestro código, modifica el tuyo de la siguiente manera:

```python
import re

email = input("What's your email? ").strip()

if re.search(".+@.+", email):
    print("Valid")
else:
    print("Invalid")
```

Nota que no nos importa cuál sea el nombre de usuario o el dominio. Lo que nos importa es el patrón. `.+` se usa para determinar si hay algo a la izquierda de la dirección de correo electrónico y si hay algo a la derecha de la dirección de correo electrónico. Ejecutando tu código, al escribir `malan@`, notarás que la entrada es considerada como `invalid` tal como esperaríamos.

Si hubiéramos usado una expresión regular `.*@.*` en nuestro código anterior, puedes visualizar esto de la siguiente manera:

```mermaid
stateDiagram-v2
    [*] --> q1: Inicio
    q1 --> q1: Cualquier carácter (.)
    q1 --> q2: Encuentra '@'
    q2 --> q2: Cualquier carácter (.)
    q2 --> [*]: Fin de cadena (Éxito)
```



Nota la representación de la máquina de estados de nuestra expresión regular. A la izquierda, el intérprete comienza a evaluar la declaración de izquierda a derecha. Una vez que llegamos a q1 o pregunta 1, el intérprete lee una y otra vez basándose en la expresión que se le entregó. Luego, el estado se cambia mirando ahora a q2 o la segunda pregunta que está siendo validada. De nuevo, la flecha indica cómo la expresión será evaluada una y otra vez basándose en nuestra programación. Luego, como lo representa el círculo doble, se alcanza el estado final de la máquina de estados.

Considerando la expresión regular que usamos en nuestro código, `.+@.+`, puedes visualizarlo de la siguiente manera:

```mermaid
stateDiagram-v2
    [*] --> q1: Inicio
    q1 --> q1: Cualquier carácter (.)
    q1 --> q2: Encuentra '@'
    q2 --> q3: Cualquier carácter (.)
    q3 --> q3: Cualquier carácter (.)
    q3 --> [*]: Fin de cadena (Éxito)
```



Nota cómo q1 es cualquier carácter proporcionado por el usuario, incluyendo 'q2' como 1 o más repeticiones de caracteres. Esto es seguido por el símbolo '@'. Luego, q3 busca cualquier carácter proporcionado por el usuario, incluyendo q4 como 1 o más repeticiones de caracteres.

Las funciones `re` y `re.search` y otras similares buscan patrones.

Continuando con nuestra mejora de este código, podríamos mejorar nuestro código de la siguiente manera:

```python
import re

email = input("What's your email? ").strip()

if re.search(".+@.+.edu", email):
    print("Valid")
else:
    print("Invalid")
```

Nota, sin embargo, que alguien podría escribir `malan@harvard?edu` y podría ser considerado válido. ¿Por qué es este el caso? ¡Podrías reconocer que en el lenguaje de la validación, un `.` significa cualquier carácter!

Podemos modificar nuestro código de la siguiente manera:

```python
import re

email = input("What's your email? ").strip()

if re.search(r".+@.+\.edu", email):
    print("Valid")
else:
    print("Invalid")
```

Nota cómo utilizamos el "carácter de escape" o `\` como una forma de considerar el `.` como parte de nuestra cadena en lugar de nuestra expresión de validación. Al probar tu código, notarás que `malan@harvard.edu` es considerado como válido, mientras que `malan@harvard?edu` es inválido.

Ahora que estamos usando caracteres de escape, es un buen momento para introducir las "cadenas crudas" (raw strings). En Python, las cadenas crudas son cadenas que no formatean caracteres especiales; en su lugar, cada carácter se toma por su valor literal. Imagina `\n`, por ejemplo. Hemos visto en una lección anterior cómo, en una cadena regular, estos dos caracteres se convierten en uno: un carácter especial de nueva línea. En una cadena cruda, sin embargo, `\n` no es tratado como `\n`, el carácter especial, sino como una sola `\` y una sola `n`. Colocar una `r` frente a una cadena le dice al intérprete de Python que trate la cadena como una cadena cruda, similar a cómo colocar una `f` frente a una cadena le dice al intérprete de Python que trate la cadena como una cadena de formato:

```python
import re

email = input("What's your email? ").strip()

if re.search(r"^.+@.+\.edu$", email):
    print("Valid")
else:
    print("Invalid")
```

Ahora nos hemos asegurado de que el intérprete de Python no tratará `\.` como un carácter especial. En su lugar, simplemente como una `\` seguida de un `.` —lo cual, en términos de expresiones regulares, significa hacer coincidir un "." literal.

¡Todavía puedes imaginar cómo nuestros usuarios podrían crearnos problemas! Por ejemplo, podrías escribir una oración como "Mi dirección de correo electrónico es malan@harvard.edu." y toda esta oración sería considerada válida. Podemos ser aún más precisos en nuestra codificación.

Resulta que tenemos más símbolos especiales a nuestra disposición en la validación:

* `^`   coincide con el inicio de la cadena
* `$`   coincide con el final de la cadena o justo antes de la nueva línea al final de la cadena

Podemos modificar nuestro código usando nuestro vocabulario añadido de la siguiente manera:

```python
import re

email = input("What's your email? ").strip()

if re.search(r"^.+@.+\.edu$", email):
    print("Valid")
else:
    print("Invalid")
```

Nota que esto tiene el efecto de buscar este patrón exacto coincidiendo con el inicio y el final de la expresión que está siendo validada. Escribir una oración como "Mi dirección de correo electrónico es malan@harvard.edu." ahora es considerado como inválido.

¡Proponemos que podemos hacerlo aún mejor! Aunque ahora estamos buscando el nombre de usuario al inicio de la cadena, el símbolo `@`, y el nombre de dominio al final, ¡podríamos escribir tantos símbolos `@` como deseemos! ¡`malan@@@harvard.edu` es considerado válido!

Podemos añadir a nuestro vocabulario de la siguiente manera:

* `[]`    conjunto de caracteres
* `[^]`   complementando el conjunto

Usando estas nuevas habilidades, podemos modificar nuestra expresión de la siguiente manera:

```python
import re

email = input("What's your email? ").strip()

if re.search(r"^[^@]+@[^@]+\.edu$", email):
    print("Valid")
else:
    print("Invalid")
```

Nota que `^` significa coincidir al inicio de la cadena. Todo el camino al final de nuestra expresión, `$` significa coincidir al final de la cadena. `[^@]+` significa cualquier carácter excepto un `@`. Luego, tenemos un `@` literal. `[^@]+\.edu` significa cualquier carácter excepto un `@` seguido de una expresión terminando en `.edu`. Escribir `malan@@@harvard.edu` ahora es considerado como inválido.

Todavía podemos mejorar más esta expresión regular. ¡Resulta que existen ciertos requisitos para lo que puede ser una dirección de correo electrónico! Actualmente, nuestra expresión de validación es demasiado complaciente. Podríamos querer permitir solo caracteres normalmente usados en una oración. Podemos modificar nuestro código de la siguiente manera:

```python
import re

email = input("What's your email? ").strip()

if re.search(r"^[a-zA-Z0-9_]+@[a-zA-Z0-9_]+\.edu$", email):
    print("Valid")
else:
    print("Invalid")
```

Nota que `[a-zA-Z0-9_]` le dice a la validación que los caracteres deben estar entre a y z, entre A y Z, entre 0 y 9 y potencialmente incluir un símbolo `_`. Al probar la entrada, encontrarás que se pueden indicar muchos posibles errores del usuario.

Afortunadamente, patrones comunes han sido integrados en las expresiones regulares por programadores trabajadores. En este caso, puedes modificar tu código de la siguiente manera:

```python
import re

email = input("What's your email? ").strip()

if re.search(r"^\w+@\w+\.edu$", email):
    print("Valid")
else:
    print("Invalid")
```

Nota que `\w` es lo mismo que `[a-zA-Z0-9_]`. ¡Gracias, programadores trabajadores!

Aquí hay algunos patrones adicionales que podemos añadir a nuestro vocabulario:

* `\d`    dígito decimal
* `\D`    no es un dígito decimal
* `\s`    caracteres de espacio en blanco
* `\S`    no es un carácter de espacio en blanco
* `\w`    carácter de palabra, así como números y el guion bajo
* `\W`    no es un carácter de palabra

<h2 id="case-sensitivity">2. Sensibilidad a Mayúsculas y Minúsculas (Case Sensitivity)</h2>



Para ilustrar cómo abordar los problemas de sensibilidad a mayúsculas y minúsculas (donde existe una diferencia entre `EDU` y `edu`), retomemos nuestro código base:

```python
import re

email = input("What's your email? ").strip()

if re.search(r"^\w+@\w+\.edu$", email):
    print("Valid")
else:
    print("Invalid")
```

### El parámetro `flags`
La función `re.search` incluye un parámetro para **flags** (banderas) que modifican el comportamiento del motor de búsqueda.

Algunas variables de banderas integradas son:
* **`re.IGNORECASE`**: Ignora si el texto está en mayúsculas o minúsculas.
* **`re.MULTILINE`**: Permite que `^` y `$` coincidan con el inicio/final de cada línea en lugar de solo el inicio/final de todo el string.
* **`re.DOTALL`**: Hace que el punto `.` coincida con cualquier carácter, incluyendo saltos de línea.



### Implementación de `re.IGNORECASE`
Podemos modificar nuestro código para que acepte correos en mayúsculas sin cambiar el patrón:

```python
import re

email = input("What's your email? ").strip()

# Agregamos re.IGNORECASE como tercer parámetro
if re.search(r"^\w+@\w+\.edu$", email, re.IGNORECASE):
    print("Valid")
else:
    print("Invalid")
```
Al ejecutar este programa con `MALAN@HARVARD.EDU`, la entrada ahora se considera válida.

---

### Manejo de Subdominios y Grupos

Considera la dirección de correo `malan@cs50.harvard.edu`. Con el código anterior, esta dirección sería **inválida** porque contiene un punto adicional que el patrón `\w+\.edu` no espera.

Para solucionar esto, podemos utilizar nuestro vocabulario de grupos:
* **`A|B`**: Ya sea A o B.
* **`(...)`**: Un grupo.
* **`(?:...)`**: Versión de grupo sin captura.

#### Grupos Opcionales
Podemos modificar el patrón para permitir un subdominio opcional:

```python
import re

email = input("What's your email? ").strip()

# (\w+\.)? indica que este grupo puede aparecer una vez o ninguna
if re.search(r"^\w+@(\w+\.)?\w+\.edu$", email, re.IGNORECASE):
    print("Valid")
else:
    print("Invalid")
```
Ahora, tanto `malan@cs50.harvard.edu` como `malan@harvard.edu` son considerados válidos.

---

### La complejidad de la validación real

Es importante notar que las ediciones realizadas hasta ahora no abarcan por completo todas las comprobaciones necesarias para garantizar un correo electrónico válido según los estándares oficiales. De hecho, esta es la expresión completa que se tendría que escribir para asegurar una validación estricta:

> `^[a-zA-Z0-9.!#$%&'*+\/=?^_`{|}~-]+@[a-zA-Z0-9](?:[a-zA-Z0-9-]{0,61}[a-zA-Z0-9])?(?:\.[a-zA-Z0-9](?:[a-zA-Z0-9-]{0,61}[a-zA-Z0-9])?)*$`



---

### Otras funciones útiles en `re`

Además de `re.search`, existen otras funciones en la librería que pueden ser sumamente útiles según el caso de uso:

* **`re.match`**: Intenta aplicar el patrón desde el **inicio** de la cadena (similar a usar `^` automáticamente).
* **`re.fullmatch`**: Requiere que **toda** la cadena coincida con el patrón (similar a usar `^` y `$` simultáneamente).

Para profundizar más, puedes consultar la [documentación oficial de Python sobre `re`](https://docs.python.org/3/library/re.html).

<h2 id="cleaning-up">3. Limpieza de la Entrada del Usuario</h2>



Nunca debes esperar que tus usuarios sigan siempre tus deseos de obtener una entrada "limpia". De hecho, los usuarios a menudo violarán tus intenciones como programador. Existen diversas formas de limpiar estos datos utilizando expresiones regulares.

### El problema del formato del nombre

Para comenzar, crea un archivo llamado `format.py` y codifica lo siguiente:

```python
name = input("What's your name? ").strip()
print(f"hello, {name}")
```

Este programa funciona bien si el usuario escribe `David`. Sin embargo, si el usuario escribe `Malan, David`, el programa no funcionará como se pretende (imprimirá "hello, Malan, David" en lugar de un saludo más natural).

### Primera aproximación: Métodos de String
Podemos intentar limpiar la entrada detectando la coma:

```python
name = input("What's your name? ").strip()
if "," in name:
    last, first = name.split(", ")
    name = f"{first} {last}"
print(f"hello, {name}")
```

Este código funciona para `Malan, David`, pero arrojará un error si el usuario escribe `Malan,David` (sin espacio), ya que `split(", ")` no encontrará la coincidencia exacta.

---

### Uso de Grupos de Captura con `re.search`

Podemos aplicar la sintaxis de expresiones regulares para extraer partes específicas de la entrada mediante **grupos de captura** (paréntesis).

```python
import re

name = input("What's your name? ").strip()
matches = re.search(r"^(.+), (.+)$", name)

if matches:
    last, first = matches.groups()
    name = first + " " + last

print(f"hello, {name}")
```



#### Acceso mediante `matches.group`
En lugar de desempaquetar todos los grupos a la vez, podemos solicitar grupos específicos por su índice (empezando por 1):

```python
import re

name = input("What's your name? ").strip()
matches = re.search(r"^(.+), (.+)$", name)

if matches:
    # group(2) es el nombre, group(1) es el apellido
    name = matches.group(2) + " " + matches.group(1)

print(f"hello, {name}")
```

---

### Flexibilización del Patrón (Espacios Opcionales)

Para manejar casos como `Malan,David` (sin espacio) o `Malan,    David` (muchos espacios), utilizamos el cuantificador `*`, que significa "0 o más repeticiones":

```python
import re

name = input("What's your name? ").strip()
# El espacio seguido de * permite que el espacio sea opcional o múltiple
matches = re.search(r"^(.+), *(.+)$", name)

if matches:
    name = matches.group(2) + " " + matches.group(1)

print(f"hello, {name}")
```

---

### El Operador Walrus (`:=`)

Es muy común realizar una búsqueda y luego verificar si hubo éxito. Podemos combinar estas dos líneas utilizando el **operador walrus** (`:=`), que asigna un valor a una variable y permite realizar una evaluación booleana al mismo tiempo:

```python
import re

name = input("What's your name? ").strip()

# El operador := asigna el resultado a 'matches' y evalúa si no es None
if matches := re.search(r"^(.+), *(.+)$", name):
    name = matches.group(2) + " " + matches.group(1)

print(f"hello, {name}")
```

> **Dato curioso**: Se llama "operador walrus" (morsa) porque si giras la cabeza hacia un lado, los dos puntos y el igual (`:=`) parecen los ojos y colmillos de una morsa.

<h2 id="extracting-input">4. Extracción de Información del Usuario (Extracting User Input)</h2>

Hasta ahora, hemos validado y limpiado la entrada del usuario. El siguiente paso lógico es aprender a **extraer** información específica de una cadena, ignorando el resto del contenido.

### El problema: Extraer el nombre de usuario de Twitter

Imagina que quieres obtener el *handle* de un usuario a partir de una URL de Twitter. Si creamos un archivo `twitter.py` básico:

```python
url = input("URL: ").strip()
print(url)
```

Si el usuario escribe `https://twitter.com/davidjmalan`, obtenemos la URL completa. ¿Cómo extraemos solo el nombre de usuario?

#### Intentos manuales: `replace` y `removeprefix`
Podemos intentar eliminar el inicio de la URL de forma literal:

```python
# Usando replace
username = url.replace("https://twitter.com/", "")

# Usando removeprefix (específico para el inicio de un string)
username = url.removeprefix("https://twitter.com/")
```

**Limitación:** Estos métodos fallan si el usuario omite el protocolo (`https://`), usa `http://` o incluye el subdominio `www.`.

---

### Sustitución con `re.sub`

La librería `re` ofrece el método `sub`, que permite sustituir un patrón por otra cosa.

**Firma:** `re.sub(pattern, repl, string, count=0, flags=0)`.



Podemos usarlo para "limpiar" la URL y dejar solo el nombre:

```python
import re

url = input("URL: ").strip()
# Sustituimos el patrón de la URL por una cadena vacía ""
username = re.sub(r"^(https?://)?(www\.)?twitter\.com/", "", url)
print(f"Username: {username}")
```

> **Análisis del patrón:**
> * `^`: Asegura que empezamos desde el principio.
> * `(https?://)?`: Hace que `http://` o `https://` sean opcionales.
> * `(www\.)?`: Hace que el subdominio `www.` sea opcional.
> * `\.`: Escapa el punto para que se trate de forma literal.

---

### Extracción Precisa con `re.search` y Grupos

Aunque `re.sub` es útil, a veces es más seguro buscar el patrón y **capturar** explícitamente el nombre de usuario.

```python
import re

url = input("URL: ").strip()
# Capturamos el final de la URL con (.+)$
matches = re.search(r"^https?://(www\.)?twitter\.com/(.+)$", url, re.IGNORECASE)

if matches:
    print(f"Username:", matches.group(2))
```



#### Optimización: Grupos sin captura y Operador Walrus
Podemos refinar el código usando `(?:...)` para agrupar sin crear un índice de captura y el operador `:=` para asignar y validar en una sola línea:

```python
import re

url = input("URL: ").strip()

# (?:www\.) es un grupo sin captura; (.+) es el grupo de captura 1
if matches := re.search(r"^https?://(?:www\.)?twitter\.com/(.+)$", url, re.IGNORECASE):
    print(f"Username:", matches.group(1))
```



---

### Validación de Caracteres Específicos

Para ser aún más estrictos y asegurar que el nombre de usuario solo contenga caracteres permitidos por la plataforma (letras, números y guiones bajos), ajustamos el grupo de captura:

```python
import re

url = input("URL: ").strip()

# [a-z0-9_]+ solo permite caracteres alfanuméricos y guiones bajos
if matches := re.search(r"^https?://(?:www\.)?twitter\.com/([a-z0-9_]+)", url, re.IGNORECASE):
    print(f"Username:", matches.group(1))
```

Con esto, garantizamos que el programa solo extraiga información válida según las reglas del dominio.

<h2 id="summing-up">5. Resumen de la Semana 7</h2>

En esta unidad hemos transformado nuestra capacidad de manipular texto. Mientras que los métodos nativos de `str` son útiles para tareas sencillas, las **expresiones regulares (Regex)** nos permiten definir lenguajes formales para describir patrones complejos dentro de los datos.

### 🧠 Conceptos Clave Dominados

* **Validación con `re.search`**: Uso de metacaracteres (`^`, `$`, `.`, `+`, `*`) para asegurar que una entrada cumple con un formato estricto.
* **Metacaracteres y Clases de Caracteres**: Diferenciación entre caracteres literales y símbolos con significado especial como `\w` (palabras), `\d` (dígitos) y `\s` (espacios).
* **Grupos de Captura `(...)`**: Técnica para aislar y extraer fragmentos específicos de una cadena (como nombres de usuario de una URL).
* **Limpieza con `re.sub`**: Sustitución dinámica de patrones para normalizar bases de datos o entradas de usuario.
* **Operador Walrus (`:=`)**: Optimización del código al asignar el resultado de una búsqueda y validarlo en una sola línea de control.
* **Banderas de Configuración**: Uso de `re.IGNORECASE` para flexibilizar la búsqueda sin complicar el patrón.

### 🔄 Flujo de Trabajo Típico con Regex

```mermaid
flowchart LR
    A[Entrada del Usuario] --> B{re.search}
    B -- No Coincide --> C[Rechazar / Error]
    B -- Coincide --> D[Extraer Grupos]
    D --> E[Procesar Datos Limpios]
```
